```python
# Nested condition
num = int(input("Enter a number"))
if num > 0:
    print("Its an positive integer")
    if num%2 == 0:
        print("Its an even number")
    else:
        print("Its an odd number")
elif num == 0:
    print("Its zero!!")
else:
    print("Its negative number")
```
